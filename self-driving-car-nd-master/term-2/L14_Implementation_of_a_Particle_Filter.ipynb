### Initialisation
* Number of sampling points
    * Want enough to cover the possible positions.
    * Too many particles slow down your filter (need to localise in real time).
* Ways of sampling
    * Uniformly (not practical if state space is too large, e.g. entire surface of the earth).
    * Sample around some initial estimate, e.g. GPS.

#### Data Association
* Problem of matching observed landmarks to landmarks on the map.
    * Sometimes you have map landmarks that have multiple corresponding lidar measurements
* Possible Solution: **Nearest Neighbour** rule: pick lidar measurement closest to map landmark.
![](images/14.5.png)
    * More effective with high signal-to-noise ratio for sensors (so you don't match spurious measurements).
    * Inefficient (O(# maplandmarks * # sensor measurements)
    * Not robust to errors in the vehicle position (shifted measurements)
    * Does not take uncertainties of the sensor into account (e.g. radar has precise bearing, less precise range)
